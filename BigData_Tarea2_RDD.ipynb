```python
# Convertimos el DataFrame 'youtube_clean' a un RDD
# Cada fila del DataFrame se convierte en un objeto Row dentro del RDD
youtube_rdd = youtube_clean.rdd

# Mostramos las primeras 5 filas del RDD para verificar que se creó correctamente
youtube_rdd.take(5)
```

```python
# Usamos map() para transformar cada fila del RDD
# Extraemos únicamente las columnas numéricas que queremos analizar
# En este caso: views, likes, dislikes y comment_count

numeric_rdd = youtube_rdd.map(lambda row: (
    row.views,
    row.likes,
    row.dislikes,
    row.comment_count
))
```

```python
import numpy as np

# Convertimos cada tupla en un arreglo de numpy
# Luego utilizamos mean() para calcular el promedio de cada columna
stats = numeric_rdd.map(lambda x: np.array(x)).mean()

# Mostramos los promedios calculados
print("Promedio views:", stats[0])
print("Promedio likes:", stats[1])
print("Promedio dislikes:", stats[2])
print("Promedio comentarios:", stats[3])
```

```python
# Usamos reduce() para comparar cada registro del RDD
# y quedarnos con el valor máximo de cada variable
import builtins
max_values = numeric_rdd.reduce(lambda a, b: (
    builtins.max(a[0], b[0]),  # máximo de views
    builtins.max(a[1], b[1]),  # máximo de likes
    builtins.max(a[2], b[2]),  # máximo de dislikes
    builtins.max(a[3], b[3])   # máximo de comentarios
))

# Mostramos los valores máximos encontrados
print("Máximo views:", max_values[0])
print("Máximo likes:", max_values[1])
print("Máximo dislikes:", max_values[2])
print("Máximo comentarios:", max_values[3])
```

```python
# Similar al cálculo de máximos, pero ahora obtenemos los valores mínimos

min_values = numeric_rdd.reduce(lambda a, b: (
    builtins.min(a[0], b[0]),  # mínimo de views
    builtins.min(a[1], b[1]),  # mínimo de likes
    builtins.min(a[2], b[2]),  # mínimo de dislikes
    builtins.min(a[3], b[3])   # mínimo de comentarios
))

# Mostramos los valores mínimos encontrados
print("Mínimo views:", min_values[0])
print("Mínimo likes:", min_values[1])
print("Mínimo dislikes:", min_values[2])
print("Mínimo comentarios:", min_values[3])
```

```python
# Extraemos únicamente la columna 'views' del RDD
views_rdd = youtube_rdd.map(lambda row: row.views)

# Utilizamos la función stats() que calcula estadísticas descriptivas básicas
views_stats = views_rdd.stats()

# Mostramos los resultados obtenidos
print("Estadísticas de views:")
print("count:", views_stats.count())      # número de registros
print("mean:", views_stats.mean())        # promedio
print("min:", views_stats.min())          # valor mínimo
print("max:", views_stats.max())          # valor máximo
print("variance:", views_stats.variance()) # varianza
```
